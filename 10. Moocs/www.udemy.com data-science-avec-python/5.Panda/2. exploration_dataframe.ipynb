```python
# Exploration du dataframe
# lire un fichier et en faire un dataframe pour gérer .
# On trouve ce bon fichier sur kaggle.com
import pandas
google_app = pandas.read_table("googleplaystore.csv",sep=",")
google_app
```

```python
# Verifier le nombre de isna dans la col rating
google_app["Rating"].isna().sum()
```

```python
# Verifier le nombre de isna dans toutes les cols
google_app.isna().sum()
```

```python
# Supprimer les lignes avec NAN /  Nettoyer le tableau
google_app_sansNan = google_app.dropna()
```

```python
google_app_sansNan
```

```python
google_app_sansNan.isna().sum()
```

```python
# Trier tableau
google_app_sansNan.sort_values(by="Rating").head()
```

```python
google_app_sansNan.describe()
```

```python
google_app_sansNan.dtypes
```

```python
# Sonder les variables qualitatives et quantitatives
google_app_sansNan.describe(include="all")
```

```python
# Remplacer des var  qualitatives en quantitatives 
# Retirer le + des vars dans la col "Installs"
google_app_sansNan["Installs"] = google_app_sansNan["Installs"].apply(lambda x: x.replace(",",""))
google_app_sansNan["Installs"] = google_app_sansNan["Installs"].apply(lambda x: x.replace("+",""))
```

```python
google_app_sansNan.head()
```

```python
# Transformer la colonne "Installs" de obj à int
google_app_sansNan["Installs"] = google_app_sansNan["Installs"].apply(lambda x: int(x))
```

```python
google_app_sansNan.dtypes
```

```python
# On a désormais de nouvelles stats
google_app_sansNan.describe()
```

```python
# Transformer la colonne "Reviews"" de obj à int
google_app_sansNan["Reviews"] = google_app_sansNan["Reviews"].apply(lambda x: int(x))
```
