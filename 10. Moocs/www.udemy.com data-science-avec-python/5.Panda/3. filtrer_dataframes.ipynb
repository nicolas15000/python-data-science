```python
# Filtrage du dataframe
# lire un fichier et en faire un dataframe pour gérer .
# On trouve ce bon fichier sur kaggle.com
import pandas
google_app = pandas.read_table("googleplaystore.csv",sep=",")
google_app
# Supprimer les lignes avec NAN /  Nettoyer le tableau
google_app_sansNan = google_app.dropna()
google_app_sansNan
```

```python
# Filtrer les apps dont le rating est supérieur à 4.5
google_app_sansNan.loc[google_app_sansNan["Rating"]>4.5,:]
```

```python
# Filtrer les apps dont la catégorie est business
google_app_sansNan.loc[google_app_sansNan["Category"] == "BUSINESS",:]
```

```python
# Combiner les filtres
google_app_sansNan.loc[(google_app_sansNan["Category"] == "BUSINESS") & (google_app_sansNan["Rating"] > 4),:] 
```
