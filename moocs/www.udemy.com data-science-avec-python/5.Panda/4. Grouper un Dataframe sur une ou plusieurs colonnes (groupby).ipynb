```python
# GroupBy sur un dataframe
# lire un fichier et en faire un dataframe pour gérer .
# On trouve ce bon fichier sur kaggle.com
import pandas
google_app = pandas.read_table("googleplaystore.csv",sep=",")
google_app
# Supprimer les lignes avec NAN /  Nettoyer le tableau
google_app_sansNan = google_app.dropna()
google_app_sansNan
```

```python
# Obtenir des infos sur un groupby()
import numpy as np  

# On rêgle l'affichage du tableau pour bien voir toutes les colonnes dans jupyterlab
# On affiche que 10 lignes
pandas.set_option('display.max_rows', 20)
pandas.set_option('display.max_columns', 500)
# Largeur de l'écran
pandas.set_option('display.width', 400)

# On fait un groupby, puis des stats sur chaque colonne, évidemment, si les colonnes n'ont pas étés nettoyées ( Oter les caractères et caster en INT par exemple),
# il ne peut pas encore calculer des stats ! On voit que la colonne price ne fonctionne pas par exemple.
google_app_sansNan.groupby("Category").describe(include ="all")  
```

```python
# ON veut que les stats soient faites sur la colonne price qui est un objet
google_app_sansNan.dtypes

# Remplacer des var  qualitatives en quantitatives 
# Retirer le $ des vars dans la col "Price"
google_app_sansNan["Price"] = google_app_sansNan["Price"].apply(lambda x: x.replace("$",""))

# Transformer la colonne "Price" de obj à int
google_app_sansNan['Price'] = google_app_sansNan['Price'].astype(float)

# On voit que désormais, la moyenne des prix fonctionnne .
google_app_sansNan.groupby("Category").describe(include ="all")  
```
