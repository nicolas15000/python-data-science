```python
import matplotlib.pyplot as plt
import numpy 
import pandas

# RÃ©aliser des stats sur un fichier de ventes de black friday .
black_friday = pandas.read_csv("BlackFriday.csv")

# Seaborn a pour avantage de prendre directement les noms de colonnes .
import seaborn as sns
sns.boxplot(x="Gender",y="Purchase",hue="Gender",data = black_friday,palette = "Set1")
plt.draw()
```

```python
# Achats par catÃ©gorie d'age ET par genre Ã  l'aide de hue
sns.boxplot(x="Gender",y="Purchase",hue="Age",data = black_friday,palette = "Set1")
plt.draw()
```

```python
# countplot
sns.countplot(x="Age",hue="Gender",data = black_friday,palette = "Set1")
plt.draw()
```

```python
# GÃ©nÃ©rer un subset alÃ©atoire d'un jeu de donnÃ©es
subset = black_friday.sample(1000)
subset
```

```python
sns.catplot(x="Age",y="Purchase",data = subset)
plt.draw()
```

```python
sns.catplot(x="Age",y="Purchase",data = subset,kind="swarm")
plt.draw()
```

```python
# On visualise le genre
sns.catplot(x="Age",y="Purchase",hue="Gender",data = subset,kind="swarm")
plt.draw()
```

```python
#Combiner des graphiques
plt.subplot(211)
sns.countplot(x="Age",hue="Gender",data = black_friday,palette = "Set1")
plt.subplot(212)
sns.boxplot(x="Gender",y="Purchase",hue="Age",data = black_friday,palette = "Set1")
plt.legend(bbox_to_anchor=(1.05,1),loc=2)
plt.draw()
```
