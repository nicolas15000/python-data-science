```python
import matplotlib.pyplot as plt
import numpy 
import pandas
import seaborn as sns

# Exercice de fin dechapitre sur seaborn
tips = sns.load_dataset('tips')
tips.columns
```

```python
tips
```

```python
# 1. Tracer un graphique reprÃ©sentant la variable total_bill, sÃ©parÃ©e par sexe
sns.barplot(y="total_bill",x="sex",data = tips)
plt.draw()
```

```python
# 2. Etudier la distribution de la variable total_bill Ã  l'aide de la fonction sns.distplot()
# Je prends toutes les valeurs de la colonne total_bill puis les mets dans une liste python
toutes_mes_vars = tips['total_bill'].to_list()
# Du coup, ensuite, je peux Ã©tudier la distribution.
sns.distplot(toutes_mes_vars)
```

```python
# 3. Utiliser la function jointplot de seaborn pour tracer la variable total_bill X en fonction de la variable tip (Y)
g = sns.jointplot(x="total_bill", y="tip", data=tips)
g
```

```python
# 4. Utiliser la function pairplot de seaborn sur l'ensemble du dataframe 'tips' pour Ã©tudier les relations sur l'ensemble des variables

g = sns.pairplot(data=tips)
g
```

```python
# 5. Tracer le boxplot de la variable total_bill, en crÃ©ant un boxplot par jour (variable day) et en utilisant la variable sex dans l'option 'hue'
sns.boxplot(data = tips, x = "total_bill",y="day" ,hue = "sex")
```
